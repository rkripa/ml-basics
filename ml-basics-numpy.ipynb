```python
# Covariance of a matrix. Its shape will be the # of featuers (columns)

a = np.random.randn(100,3)  # 100 samples of each 3 features
np.cov(a).shape
```

```python
np.cov(a.T).shape
```

```python
cov = np.cov(a.T)
cov
```

```python
# Covariance is a symteric matrix so we can use eigh
np.linalg.eigh(cov)
```

```python
np.linalg.eig(cov)
```
