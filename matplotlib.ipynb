```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# linspace takes starting, ending and the # points inbetween
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)

plt.xlabel('time')
plt.ylabel('amplituude')
plt.title('Sine')
plt.show()
```
