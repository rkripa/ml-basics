```python
# Sampling from Gaussian
g = np.random.randn(10000)
plt.hist(g, bins=100)
plt.show()
```

```python
# Std deviation 10 and mean five
# We scale it by 10 (sd) and add 5 (mean)
g = 10* np.random.randn(10000) + 5
plt.hist(g, bins=100)
plt.show()
```

```python
g.mean(), g.std()
```

```python
# Sample from 2d gaussian mean 0, variance 1.  
# This is called sperical gaussian.  
# Each dimension is uncorrelated and independent of the other
# They have the same variance

g = np.random.randn(10000,200)
plt.scatter(g[:,0], g[:,1])
plt.axis('equal')
plt.show()
```

```python
# Elliptical Gaussian. The variance is different in different dimensions
# Lets scale for std dev of 5 and mean of 2
g[:,1] = 5*g[:,1] + 2
plt.scatter(g[:,0], g[:,1])
plt.show()
```

```python
# Plot with equal axis
plt.scatter(g[:,0], g[:,1])
plt.axis('equal')
plt.show()
```

Is it possible to create ellipses not aligned with the axises.
This would mean a full covariance matrix with non zero in off diagonal.
Dimensions would be correlated with each other.
How to sample from a multivariate normal distributions where the dimensions are not necessarily independent from one another. This is a full covariance matrix.

Variance of 1 in the first dimension and a variance of 3 in the second dimension. The covariance between the 2 dimensions is 0.8

```python
cov = np.array([[1,0.8], [0.8, 3]])
```

```python
from scipy.stats import multivariate_normal as mvn
```

```python
mu = np.array([0,2])
r = mvn.rvs(mean=mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

numpy.ffe has functions of FFT.  Convert from time domain to frequency domain

```python
x = np.linspace(0,100,10000)
y = np.sin(x) + np.sin(3*x) + np.sin(5*x)
plt.plot(y)
plt.show()
```

```python
Y = np.fft.fft(y)
plt.plot(np.abs(Y))
plt.show()

```
